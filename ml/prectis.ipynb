```python
#image embozing
import cv2
import matplotlib.pyplot as plt 
import numpy as np

img=cv2.imread('aa.png')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

k1=np.array([[0,-1,-1],[1,0,-1],[1,1,0]])
k2=np.array([[-1,-1,0],[-1,0,1],[0,1,1]])
k3=np.array([[1,0,0],[0,0,0],[0,0,-1]])
k4=np.array([[0,0,1],[0,0,0],[-1,0,0]])

img1=cv2.filter2D(img,-1,k1)
img2=cv2.filter2D(img,-1,k2)
img3=cv2.filter2D(img,-1,k3)
img4=cv2.filter2D(img,-1,k4)

plt.subplot(2,3,1),plt.imshow(img),plt.title('source'),plt.axis('off')
plt.subplot(2,3,2),plt.imshow(img1,cmap='gray'),plt.title('k1'),plt.axis('off')
plt.subplot(2,3,3),plt.imshow(img2,cmap='gray'),plt.title('k2'),plt.axis('off')
plt.subplot(2,3,4),plt.imshow(img3,cmap='gray'),plt.title('k3'),plt.axis('off')
plt.subplot(2,3,5),plt.imshow(img4,cmap='gray'),plt.title('k4'),plt.axis('off')
```

```python
#erostion and dilation
import cv2
import numpy as np
import matplotlib.pyplot as plt

#load the image
img=cv2.imread('11.webp')
img=cv2.bitwise_not(img)
#defining a kernal
k1=np.ones((3,3),np.uint8)

#erostion
erostion=cv2.erode(img,k1,iterations=1)

#ditation
dilation=cv2.dilate(img,k1,iterations=1)

#display the imng
#plt.subplot(1,3,1),plt.imshow(img,cmap='gray'),plt.title("source")
#plt.subplot(1,3,2),plt.imshow(erostion,cmap='gray'),plt.title("erostion")
#plt.subplot(1,3,3),plt.imshow(dilation,cmap='gray'),plt.title("dilation")

cv2.imshow('Source',img)
cv2.imshow('Erostion',erostion)
cv2.imshow('Dilation',dilation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#contrasting the color image
#using the global histrogram equivalance(ghe)
import cv2
import numpy as np
import matplotlib.pyplot as plt

#loading the image
img=cv2.imread('rose.jpg')

#converting the image from bgr to ycrcb
ycrcb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
y,cr,cb=cv2.split(ycrcb)

#applying the color from ycrcb to bgr
y_eq=cv2.equalizeHist(y)
ycrcb_eq=cv2.merge([y_eq,cr,cb])

#converting the color from ycrcb to bgr
global_eq=cv2.cvtColor(ycrcb_eq,cv2.COLOR_YCrCb2BGR)

#converting the color from bgr to rgb
global_eq1=cv2.cvtColor(global_eq,cv2.COLOR_BGR2RGB)
source=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#ploting the image
plt.subplot(1,2,1),plt.imshow(source,cmap='gray'),plt.title('source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(global_eq1,cmap='gray'),plt.title('ghe img'),plt.axis('off')
```

```python
#contrasting the color image
#using the clahe
import cv2
import numpy as np
import matplotlib.pyplot as plt

#loading the image
img=cv2.imread('p.jpg')

#converting the image from bgr to ycrcb
ycrcb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
#split the elemrnt of the img
y,cr,cb=cv2.split(ycrcb)

#applying the contrast using clahe
clahe=cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
y_clahe=clahe.apply(y)
ycrcb_clahe=cv2.merge([y_clahe,cr,cb])

#converting the color from ycrcb to bgr
global_clahe=cv2.cvtColor(ycrcb_clahe,cv2.COLOR_YCrCb2BGR)

#converting the color from bgr to rgb
global_clahe1=cv2.cvtColor(global_clahe,cv2.COLOR_BGR2RGB)
source=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#ploting the image
plt.subplot(1,2,1),plt.imshow(source,cmap='gray'),plt.title('source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(global_clahe1,cmap='gray'),plt.title('clahe img'),plt.axis('off')
```

```python
#contrasting the color image
#using the clahe
import cv2
import numpy as np
import matplotlib.pyplot as plt

#loading the image
img=cv2.imread('p.jpg')

#converting the image from bgr to ycrcb
ycrcb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
#split the elemrnt of the img
y,cr,cb=cv2.split(ycrcb)

#applying the contrast using clahe
clahe=cv2.createCLAHE(clipLimit=3.0,tileGridSize=(8,8))
y_clahe=clahe.apply(y)
ycrcb_clahe=cv2.merge([y_clahe,cr,cb])

#converting the color from ycrcb to bgr
global_clahe=cv2.cvtColor(ycrcb_clahe,cv2.COLOR_YCrCb2BGR)

#converting the color from bgr to rgb
global_clahe1=cv2.cvtColor(global_clahe,cv2.COLOR_BGR2RGB)
source=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#ploting the image
plt.subplot(1,2,1),plt.imshow(source,cmap='gray'),plt.title('source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(global_clahe1,cmap='gray'),plt.title('clahe img'),plt.axis('off')
```

```python
#appliying advance image filters
#1)bilateral
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('ff.jpg')
bilateral=cv2.bilateralFilter(img,d=150,sigmaColor=150,sigmaSpace=150)

plt.subplot(1,2,1),plt.imshow(img),plt.title('source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(bilateral),plt.title('smooth image using bilateral'),plt.axis('off')
```

```python
#appliying advance image filters
#1)bilateral
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('mount.jpg')

#blur
gauss=cv2.GaussianBlur(img,(9,9),10)
#unsharp masking
unsharp=cv2.addWeighted(img,1.5,gauss,-0.5,0)

plt.subplot(1,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.title('source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(cv2.cvtColor(unsharp,cv2.COLOR_BGR2RGB)),plt.title('sharpning  using unsharp'),plt.axis('off')
```

```python
#scharr (for edge detection)
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('rose.jpg',0)

#apply scharr filter
scharrx=cv2.Scharr(img,cv2.CV_64F,1,0)
scharry=cv2.Scharr(img,cv2.CV_64F,0,1)

#convert to uint8 for display
scharrx=cv2.convertScaleAbs(scharrx)
scharry=cv2.convertScaleAbs(scharry)

#combination both direction
scharr_combine=cv2.addWeighted(scharrx,0.5,scharry,0.2,0)

#ploat the image
plt.subplot(2,2,1),plt.imshow(img,cmap='gray'),plt.title('scource'),plt.axis('off')
plt.subplot(2,2,2),plt.imshow(scharrx,cmap='gray'),plt.title('scharrx'),plt.axis('off')
plt.subplot(2,2,3),plt.imshow(scharry,cmap='gray'),plt.title('scharry'),plt.axis('off')
plt.subplot(2,2,4),plt.imshow(scharr_combine,cmap='gray'),plt.title('scharr_combine'),plt.axis('off')
```

```python
# Object detection from the image
# Face detection using Haar Cascade

import cv2

# Load pre-trained face detector (Haar Cascade)
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Load image
img = cv2.imread('perso.jpg')
gray_scale=(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY))#haar cascade works on single chanel color.

# Detect faces in the image
faces = face_cascade.detectMultiScale(gray_scale, scaleFactor=1.1, minNeighbors=4)

print(faces) 

# Draw rectangles around detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 3)

cv2.imshow('Face Detector', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Object detection using camera
#Face
import cv2
#Load pre-trained face detector (Haar Cascade)
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
#Open the camera (0 for inbuilt camera, 1 for external camera)
cap=cv2.VideoCapture(0)
#Capture from the camera
while True:
    #Capture frame by frame
    ret,frame=cap.read()
    if not ret:
        break #If camera not working
gray=(cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)) #Haar cascade works on single channel color.
    #Detect faces
faces=face_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=4)
print(faces)
    # Draw rectangles around faces
for(x,y,w,h) in faces:
    cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),3)
cv2.imshow('Face Detection from camera',frame)
    #Close the camera when Esc key is pressed

 
if cv2.waitKey(1)==27: #27 is the ASCII value of Esc
    break
cv2.destroyAllWindows()


```
